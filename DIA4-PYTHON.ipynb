## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida

print(mensaje_completo("Kevin"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabor disponible", ",", sabores)
  # lower() hace minusculas las palabras
  #strip() elimina los espacios tanto antes como despues
  sabor = input("Elige un sabor:").strip().lower()
  #si sabor NO esta en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente):").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
    cantidad = int(input("Cuantas bolas de nieve?(Ingresa un valor valido): "))
  return cantidad
# Funcion Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant =  pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderas? :"))
for i in range(total_pedidos):
  pedir_helado()
```

## Funciones Lambda
Una funcion lambda es una funcion acortada para escribir una funcion de esta forma:

nombreFuncion = lambda parametros: accion

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco mas limitadas

```python
sumar = lambda x,y : x + y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio : ((precio * (1-desc)),(precio * desc))
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
def promediosYmayor(promedios):
  promedio = 0
  for x in promedios:
    promedio += x
  promedio = promedio/len(promedios)
  return promedio, max(promedios)

promedios = []
continuar = ""

while True:
  if len(promedios) > 0:
    continuar=""
    while continuar not in ("si", "no"):
      continuar = input("Ingrese una respuesta válida: si/no").lower()
    if continuar == "no":
        break

  while True:
      try:
          calif = int(input("Indique la calificación: "))
          if calif < 0 or calif > 100:
              print("Ingrese un valor valido")
          else:
              promedios.append(calif)
              break
      except ValueError:
          print("Ingrese un valor numérico")

prom, mayor = promediosYmayor(promedios)

print("\nResultados:")
print("Promedio:", prom)
print("Mayor:", mayor)
```

```python
# Calculadora -> suma, resta, producto, division y potencia
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def producto(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Error: no se puede dividir entre 0"
    return a / b

def potencia(a, b):
    return a ** b

def menu():
    print("-------- MI CALCULADORA --------")
    print("1. Suma")
    print("2. Resta")
    print("3. Producto")
    print("4. División")
    print("5. Potencia")
    print("6. Salir")

while True:
    menu()
    try:
        op = int(input("Seleccione una opción: "))
    except ValueError:
        print("Ingrese un número válido")
        continue

    if op == 6:
        print("SALIENDO...")
        break

    if op < 1 or op > 6:
        print("Opción inválida")
        continue

    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
    except ValueError:
        print("Error: Ingrese un número válido")
        continue

    if op == 1:
        print("Resultado:", suma(a, b))
    elif op == 2:
        print("Resultado:", resta(a, b))
    elif op == 3:
        print("Resultado:", producto(a, b))
    elif op == 4:
        print("Resultado:", division(a, b))
    elif op == 5:
        print("Resultado:", potencia(a, b))
```
