## Operaciones matematicas

```python
# 4to Nivel de Importancia
# Suma Ordinaria
print("Suma ordinaria: ")
print(3 + 4)
variable1,variable2 = 20,40 # 60
print(variable1 + variable2)

# Resta Ordinaria
print("Resta Ordinaria: ")
print(2 - 5) # -3
# Declaracion de variables
v1 = 10.4
v2 = 15
resultado = v2 - v1
print(resultado) # 4.6
#Phyton es el lenguaje predilecto para absolutamente todos los procesos matematicos
```

```python
# 3er Nivel de importancia
# Multiplicacion ordinaria
print("Multiplicacion Ordinaria: ")
M1, M2 = 2,3.14
print(M1 * M2) # 6.28
#Division Ordinaria
print("-----------------")
print("Division Ordinaria: ")
M1,M2,M3,M4,M5,M6 = 2,5,6,5,4,5
print(M1/M2)
print(M3/M4)
print(M5/M6)
# Divison de Piso
print("-----------------")
print("Division de Piso")
print(M1//M2)
print(M3//M4)
print(M5//M6)

print( 6 / 2 * 3)
```

```python
# Potencias
print("Potencias")
print(2 ** 3) # 8
P1,P2 = 7,2
print(P1 ** P2) # 49
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
A
```
